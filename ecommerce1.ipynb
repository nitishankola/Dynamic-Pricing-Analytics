```python
# 1) Basic cleaning pass
df.drop_duplicates()
```

```python
# numeric conversions
num_cols = ["cost_price","base_price","actual_price","discount_percent","units_sold","demand_score",
            "competitor_price","page_views","add_to_cart_rate","conversion_rate","stock_available",
            "lead_time_days","price_elasticity"]

for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")
```

```python
# Flags and fixes
df["price_below_cost"] = (df["actual_price"] < df["cost_price"]).astype(int)

df.loc[df['add_to_cart_rate'] > 1 ,'add_to_cart_rate'] = np.nan
df.loc[df["conversion_rate"] > 1, "conversion_rate"] = np.nan

# Impute selected columns but keep missing flags
```

```python
for col in ["units_sold","demand_score","page_views"]:
    if col in df.columns:
        df[f"{col}_missing"] = df[col].isna().astype(int)
        df[col] = df[col].fillna(df[col].median())
```

```python
df.columns
```

# Feature engineering

```python
df['revenue'] = df['actual_price'] * df['units_sold']
df['margin'] = df['actual_price'] - df['cost_price']
df["gross_margin_pct"] = df['margin']/df['actual_price'].replace(0, np.nan)
df["price_gap"] =df['actual_price'] -df['competitor_price']
df["discount_amount"] = df["base_price"] - df["actual_price"]
```

```python
df["month"] = df['date'].dt.month
```

```python
df['date'] = pd.to_datetime(df['date'])
df["week_of_year"]  = df['date'].dt.isocalendar().week
df["is_weekend"]  = (df['date'].dt.weekday>=5).astype(int)

df["stock_bucket"] = pd.cut(df['stock_available'],bins=[-1,10,100,99999], labels=["Low","Medium","High"])
```

```python
df["is_promotional"] = (df["discount_percent"] > 5).astype(int)
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf



# Weekly aggregation
df_week = df.groupby([pd.Grouper(key="date", freq="W"), "category"]).agg(
    units_sold=("units_sold","sum"),
    actual_price=("actual_price","mean"),
    page_views=("page_views","sum"),
    discount_percent=("discount_percent","mean"),
    competitor_price=("competitor_price","mean")
).reset_index()

# Log transforms
df_week["log_units"] = np.log(df_week["units_sold"] + 1)
df_week["log_price"] = np.log(df_week["actual_price"])

results = []

for cat, grp in df_week.groupby("category"):
    try:
        model = smf.ols(
            'log_units ~ log_price + np.log(page_views + 1) + discount_percent',
            data=grp
        ).fit()

        beta = model.params.get("log_price", np.nan)
        pval = model.pvalues.get("log_price", np.nan)

        results.append({
            "category": cat,
            "elasticity": beta,
            "pvalue": pval
        })

        print(f"Category: {cat}, Elasticity: {beta:.3f}, P-value: {pval:.3f}")

    except Exception as e:
        print("Error for", cat, e)
        results.append({
            "category": cat,
            "elasticity": np.nan,
            "pvalue": np.nan
        })

# Save output
res_df = pd.DataFrame(results).sort_values("elasticity")
res_df.to_csv("category_elasticity.csv", index=False)

print(" Saved category_elasticity.csv")
```
